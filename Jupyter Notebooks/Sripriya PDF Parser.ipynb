```python
%matplotlib inline
%run ./plugins/widgets.py

import glob
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pdf2image
import pickle
import PIL
import PIL.Image
import pytesseract
import random
import re
from wordcloud import WordCloud, STOPWORDS
```

### Global Parameters

```python
plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['font.size'] = 12
PDF_DIR = '/Users/akshaylaradhaya/Documents/Sripriya PDF Parser Data/PDF Files'
IMG_DIR = '/Users/akshaylaradhaya/Documents/Sripriya PDF Parser Data/Title Page Images'

corrupted_files = ['1959tn1.pdf', '1961gujarat27.pdf', '1971tn15.pdf', '2003rajasthan14.pdf', '2011mp43.pdf', '2014uk5.pdf' ]
different_files = ['1959hr27.pdf', '1961pn8.pdf', '1963kr24.pdf', '1969kr1.pdf', '1993or11.pdf', '2005hr13.pdf', '2015meghalaya11.pdf']
```

### Load and Filter PDF files

```python
pdf_files = os.listdir(PDF_DIR)
if '.DS_Store' in pdf_files:
    pdf_files.remove('.DS_Store')
print("Total Number of Files  : {:4d}".format(len(pdf_files)))

# Filter Corrupted files
pdf_files = [pdf for pdf in pdf_files if pdf not in corrupted_files]
print("Corrupted Files        : {:4d}".format(len(corrupted_files)))

# Filter Non working Files
pdf_files = [pdf for pdf in pdf_files if pdf not in different_files]
print("Different Format Files : {:4d}".format(len(different_files)))

pdf_files.sort()
print("Final Number of Files  : {:4d}".format(len(pdf_files)))
```

### Generate images of title page

```python
for pdf_file in log_progress(pdf_files, name='PDF'):
    try:
        pdf_name, _ = pdf_file.split('.')
        file_path = os.path.join(PDF_DIR, pdf_file)
        img_path = os.path.join(IMG_DIR, pdf_name + '.png')
        
        if not os.path.exists(img_path):
            img = pdf2image.convert_from_path(file_path, dpi=600, first_page=1, last_page=1)[0]
            width, height = img.size
            newheight = int(height * 5000 / width)
            img = img.resize((5000, newheight))
            area = (500, 1000, 4500, 4500)
            cropped_img = img.crop(area)
            cropped_img.save(img_path, 'PNG')
    except Exception as e:
        print(str(e))
        print("Failed", pdf_file)
```

### Get OCR data

```python
cached_filename = './data/ocr_data.pkl'

if os.path.exists(cached_filename):
    with open(cached_filename, 'rb') as file:
        ocr_data = pickle.load(file)
        print("Loaded OCR data from file")
else:
    ocr_data = []
    img_files = os.listdir(IMG_DIR)
    if '.DS_Store' in img_files:
        img_files.remove('.DS_Store')
    img_files.sort()
    print("Number of Files : ", len(img_files))

    for img_file in log_progress(img_files, name='Images'):
        try:
            img_name, _ = img_file.split('.')
            file_path = os.path.join(IMG_DIR, img_file)

            img = PIL.Image.open(file_path)

            text = pytesseract.image_to_string(img)
            ocr_data.append({
                'filename': img_name,
                'text': text
            })
        except Exception as e:
            print(str(e))
            print("Failed", img_file)
            
    with open(cached_filename, 'wb') as file:
        pickle.dump(ocr_data, file)
        print("Saved OCR data to file")
```

### Parse OCR and create DataFrame

```python
df = pd.DataFrame(columns=['filename','Title','Keywords', 'Act No.', 'State', 'Year'])
filename_to_state = {
    'ap' : 'Andhra Pradesh',
    'ap' : 'Arunachal Pradesh',
    'assam' : 'Assam',
    'bihar' : 'Bihar',
    'cg' : 'Chhattisgarh',
    'delhi' : 'Delhi',
    'goa' : 'Goa',
    'gujarat' : 'Gujarat',
    'hr' : 'Haryana',
    'hp' : 'Himachal Pradesh',
    'jk' : 'Jammu and Kashmir ',
    'jr' : 'Jharkhand',
    'kr' : 'Karnataka',
    'kerala' : 'Kerala',
    'mp' : 'Madhya Pradesh',
    'mh' : 'Maharashtra',
    'manipur' : 'Manipur',
    'meghalaya' : 'Meghalaya',
    'mizoram' : 'Mizoram',
    'nl' : 'Nagaland',
    'or' : 'Orrisa',
    'pondicherry': 'Pondicherry',
    'pn' : 'Punjab',
    'rajasthan' : 'Rajasthan',
    'sikkim' : 'Sikkim',
    'sk' : 'Sikkim',
    'tn' : 'Tamil Nadu',
    'tripura' : 'Tripura',
    'up' : 'Uttar Pradesh',
    'uk' : 'Uttarakhand',
    'wb' : 'West Bengal'
}

for data in log_progress(ocr_data, name='Row'):
    
    row_data = [data['filename'], None, None, None, None, data['filename'][:4] ]
    
    text = data['text']
    lines = text.split('\n')
    lines = list(map(str.strip, lines))
    pattern = '^(Act )?([0-9]{1,2} )?of [0-9]{4}$'
    
    line_found = -1
    for i, line in enumerate(lines):
        if re.search(pattern, line):
            line_found = i
            break
        
    if line_found != -1:
        
        # Find Title
        row_data[1] = ' '.join(lines[:line_found])
        
        # Find Keywords
        row_data[2] = ''
        for line in lines[line_found+3:]:
            if line != '':
                    row_data[2] += line + ' '
        
        row_data[2] = list(map(str.strip, filter(None, row_data[2].split(','))))
    
        # Find Act No
        m = re.match(r'.*([0-9]+) of [0-9]{4}', lines[line_found])
        if m is not None:
            row_data[3] = m.group(1)
            
        # Find state
        m = re.match(r'[0-9]{4}([a-z]+)[0-9]*', data['filename'])
        if m is not None:
            abbr = m.group(1)
            if abbr in filename_to_state:
                row_data[4] = filename_to_state[abbr]
            else:
                print("Cant find state for ", data['filename'])
        
    
    df.loc[len(df)] =  row_data
    
df.to_csv('./data/pdf_details.csv')
print("Shape :", df.shape)
df.sample(10)
```

### Title word cloud

```python
all_titles = df['Title'].values
full_string = ' '.join(all_titles)
full_string = full_string.replace('Guj ', 'Guj')
full_string = full_string.replace(',', ' ')
full_string = full_string.replace('(', ' ')
full_string = full_string.replace(')', ' ')
full_string = ' '.join(full_string.split())
wordcloud = WordCloud(width=1280, height=720, stopwords=STOPWORDS, collocations = False).generate(full_string)
plt.axis("off")
_ = plt.imshow(wordcloud, interpolation='bilinear')
```
